Escrever um algoritmo para exibir os m√∫ltiplos de 3 compreendidos no intervalo: [3 100].

## Exerc√≠cio 5:
Escreva um algoritmo que leia de 10.000 habitantes de uma pequena cidade se est√° empregado ou n√£o e
exiba em porcentagem a quantidade de empregados e desempregados desta pequena cidade.


```
E = 0
D = 0
Hab = 4
for contador in range (0,Hab):
  Emp = int(input('Insira 1 para empregado e 0 para desempregado: '))
  while (Emp != 1 and Emp != 0):
    Emp = int(input('Valor INVALIDO. Insira 1 para empregado e 0 para desempregado: '))
  if (Emp == 1):
    E += 1
  else:
    D += 1

PE = (E / Hab) * 100
PD = (D / Hab) * 100
print(f'Porcentagem de empregados: {PE}%')
print(f'Porcentagem de desempregados: {PD}%')

  
```

## Exerc√≠cio 6:
Escreva um algoritmo que leia o sal√°rio em reais (R$) de 1000 clientes de um shopping e exiba na tela,
em porcentagem, a divis√£o dos clientes por tipo: A, B ou C, conforme a seguir:

image.png

```
Clie = 1000
SM = 998.05
A = 0
B = 0
C = 0
for contador in range (0,Clie):
  Sal = float(input('Insira o salario em reais: '))
  if (Sal >= (SM*15)):
    A += 1
  elif (Sal < (SM*15) and Sal >= (SM*5)):
    B += 1
  elif (Sal < (SM*5)):
    C += 1

PA = (A / Clie) * 100
PB = (B / Clie) * 100
PC = (C / Clie) * 100
print(f'Porcentagem de clientes A: {PA: .1f}%')
print(f'Porcentagem de clientes B: {PB: .1f}%')
print(f'Porcentagem de clientes C: {PC: .1f}%')


```
